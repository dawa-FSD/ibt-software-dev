```python

# Pharmacy Inventory Tracker

stock = {}

try:
    with open("stock.txt", "r") as f:
        for line in f:
            item, qty = line.strip().split(",")
            stock[item] = int(qty)

except FileNotFoundError:
    print("No stock file yet — starting empty")

def adjust(item, amount):
    stock[item] = stock.get(item, 0) + amount

def show_stock():
    print("\nCurrent Stock:")
    for item, qty in stock.items():
        print(f"{item}: {qty}")
        
def low_stock():
    low = [item for item, qty in stock.items() if qty < 10]
    print("\nLow stock:", low)

def save_stock():
    with open("stock.txt", "w") as f:
        for item, qty in stock.items():
            f.write(f"{item},{qty}\n")

while True:
    print("\n1. Show Stock")
    print("2. Add/Adjust Stock")
    print("3. Check Low Stock")
    print("4. Save and Exit")

    choice = input("Choose option: ")

    if choice == "1":
        show_stock()

    elif choice == "2":
        item = input("Enter medicine name: ")

        try:
            amount = int(input("Enter quantity to add: "))
            adjust(item, amount)
            print("Stock updated.")

        except ValueError:
            print("Quantity must be a number.")

    elif choice == "3":
        low_stock()

    elif choice == "4":
        save_stock()
        print("Stock saved. Goodbye!")
        break

    else:
        print("Invalid option.")
```
